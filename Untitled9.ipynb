## for

```python
dersad="ileri bgs prog"
for harf in dersad:
    print(harf)
```

```python
dersad="ileri bgs prog"
for harf in dersad:
    print(harf,end=" ")
```

```python
dersad="ileri bgs prog"
for i in range(len(dersad)):
    print(i,end=" ")
```

```python
i=0
while i<len(dersad):
    print(dersad[i],end=" ")
    i+=1
```
